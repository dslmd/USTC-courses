```python
from matplotlib import pyplot as plt #用来绘制图形
import numpy as np  #用来处理数据
```

```python
#打开x,y,z数据文件并转换成列表
f_x = open('1.5_Up_ELS.txt')
s_x = f_x.read()
p_x = list(float(i) for i in s_x.split())
```

```python
x = p_x[0::3]
x = np.array(x)
x += 13.348
y = p_x[1::3]
```

```python
#计算本底
number = 1157
total = 0
for i in range(number):
    total += y[i]
aver = total / number
print("截止到number=%d的本底约为："%(number)+str(aver))
```

```python
plt.figure(dpi=200)
plt.plot(x,y)
plt.xlabel("Voltage/V")
plt.ylabel("N")
plt.annotate('Peak1',xy=(21.218, 3420),xytext=(21.218+1, 3400),weight='bold',color='r',arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='r'));
plt.annotate('Peak2',xy=(23.038, 1900),xytext=(23.038+1, 1900),weight='bold',color='r',arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='r'));
plt.annotate('Peak3',xy=(23.678, 1000),xytext=(23.678+1, 1000),weight='bold',color='r',arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='r'));
plt.annotate('Peak4',xy=(23.978, 500),xytext=(23.978+1, 500),weight='bold',color='r',arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='r'));
#plt.axhline(y=aver, ls='--', c='r')
plt.title("Distribution plot with angle = 1.5")
```
