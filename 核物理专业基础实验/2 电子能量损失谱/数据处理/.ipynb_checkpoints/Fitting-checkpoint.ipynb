```python
from matplotlib import pyplot as plt #用来绘制图形
import numpy as np  #用来处理数据
from mpl_toolkits.mplot3d import Axes3D  #用来给出三维坐标系。
import math #用来转换phi，theta画图
```

```python
#打开x,y,z数据文件并转换成列表
f_x = open('1.5_Up_ELS.txt')
s_x = f_x.read()
p_x = list(float(i) for i in s_x.split())
```

```python
x = p_x[0::3]
y = p_x[1::3]
```

```python
#计算本底
number = 1100
total = 0
for i in range(number):
    total += y[i]
aver = total / number
print("截止到number=%d的本底约为："%(number)+str(aver))
```

```python
plt.figure(dpi=200)
plt.plot(x,y)
plt.xlabel("Voltage/V")
plt.ylabel("N")
plt.axhline(y=aver, ls='--', c='r')
plt.title("Distribution plot")
```
